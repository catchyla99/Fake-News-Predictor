```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake= pd.read_csv('/Fake.csv')
data_true= pd.read_csv('/True.csv')
#data imported
```

```python
data_fake.head()
```

```python
data_true.head()
```

```python
data_fake["class"]=0
data_true["class"]=1
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing = data_fake.tail(10)
for i in range(23450,23440,-1):
    data_fake.drop([i], axis=0, inplace=True)

data_true_manual_testing = data_true.tail(10)
for i in range(21396,21386,-1):
    data_true.drop([i], axis=0, inplace=True)
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing['class'] = 0
data_true_manual_testing['class'] = 1
```

```python
data_fake_manual_testing.head(10)
```

```python
data_true_manual_testing.head(10)
```

```python
data_merge = pd.concat([data_fake, data_true], axis= 0 )
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data= data_merge.drop(['title', 'subject', 'date'], axis=1)
```

```python
#check for null data
data.isnull().sum()
```

```python
#merging data randomly
data= data.sample(frac=1)
data.head()
```

```python
data.reset_index(inplace=True)
data.drop(['index'], axis=1, inplace=True)
```

```python
data.columns
```

```python
#function to process the text
#many filters apllied to the original text
def wordopt(text):
  text=text.lower()
  text= re.sub('\[.*?\]', '', text)
  text= re.sub("\\W", " ", text)
  text= re.sub('https?://\S+|www\.\S+', '', text)
  text= re.sub('<.*?>+', '', text)
  text= re.sub('[%s]' % re.escape(string.punctuation), '', text)
  text= re.sub('\n', '', text)
  text= re.sub('\w*\d\w*', '', text)
  return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

Using wordopt function on whole data

```python
x = data['text']
y= data['class']
```

```python
#split the data in training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
prediction_lr= LR.predict(xv_test)
```

```python
#accuracy score
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, prediction_lr))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train);
```

```python
prediction_dt= DT.predict(xv_test)
```

```python
#Accuracy
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, prediction_dt))
```

Gradient Boosting Clasifier

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(xv_train, y_train)
```

```python
predict_gb= GBC.predict(xv_test)
```

```python
GBC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, predict_gb))
```

Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train, y_train)
```

```python
predict_rf= RFC.predict(xv_test)
```

```python
RFC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, predict_rf))
```

Now testing the model

```python
def output_lable(n):
  if n==0:
    return "Fake News"
  elif n==1:
    return "True News"

def manual_testing(news):
  testing_news = {"text" : [news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  prediction_lr = LR.predict(new_xv_test)
  prediction_dt = DT.predict(new_xv_test)
  prediction_gb = GBC.predict(new_xv_test)
  prediction_rf = RFC.predict(new_xv_test)

  print("Logistic Regression Prediction: {}".format(output_lable(prediction_lr[0])))
  print("Decision Tree Prediction: {}".format(output_lable(prediction_dt[0])))
  print("Gradient Boosting Prediction: {}".format(output_lable(prediction_gb[0])))
  print("Random Forest Prediction: {}".format(output_lable(prediction_rf[0])))
```
